```python
def return_point(local,my_local):
 
    point=1
    
    if local==0:
        # 골목길 0
        point=point+1
        if my_local==1 or my_local==2 or my_local==5 or my_local==6:
            point=point+1
    elif local==1:
        # 절 1
        point=point+1
        if my_local==0 or my_local==2 or my_local==3:
            point=point+1
    elif local==2:
        # 번화가 2
        point=point+1
        if my_local==0 or my_local==1 or my_local==3 or my_local==6:
            point=point+1        
    elif local==3:
        # 연못 3
        if my_local==1 or my_local==2 or my_local==7 or my_local==4:
            point=point+1   
        
    elif local==4:
        # 병원 4
        if my_local==3 or my_local==7 or my_local==8:
            point=point+1   
        point=point+1
        
    elif local==5:
        # 양궁장 5
        if my_local==0 or my_local==9 or my_local==6:
            point=point+1   
        
    elif local==6:
        # 학교 6
        if my_local==0 or my_local==2 or my_local==5 or my_local==9 or my_local==10:
            point=point+1   
        
    elif local==7:
        # 묘지 7
        if my_local==3 or my_local==4 or my_local==8 or my_local==11:
            point=point+1   
        
    elif local==8:
        # 공장 8
        if my_local==4 or my_local==7 or my_local==11 or my_local==14:
            point=point+1   
        
    elif local==9:
        # 호텔 9
        point=point+1
        if my_local==5 or my_local==6 or my_local==12 or my_local==10:
            point=point+1   
        
    elif local==10:
        # 숲 10
        if my_local==2 or my_local==6 or my_local==9 or my_local==12 or my_local==13 or my_local==11:
            point=point+1   
        
    elif local==11:
        # 성당 11
        point=point+1
        if my_local==10 or my_local==7 or my_local==13 or my_local==14 or my_local==8:
            point=point+1   
        
    elif local==12:
        # 모래사장 12
        if my_local==9 or my_local==10or my_local==13:
            point=point+1   
        
    elif local==13:
        # 고급 주택가 13
        point=point+1
        if my_local==14 or my_local==11 or my_local==10or my_local==12:
            point=point+1   
        
    elif local==14:
        # 항구 14 
        if my_local==13 or my_local==11 or my_local==8:
            point=point+1    
    
     
    return point 

 
```
