```python
import pandas as pd
import random 
data = pd.read_csv('./file/ti2.txt', sep = '\n')


def line(user_line):
#     print(user_line)
#     print('count : '+str(count))
    line=user_line[int(0)].strip()
#     print(line)
    if line=='탑':
        line_point=0
        
    elif line=='정글':
        line_point=1
        
    elif line=='미드':
        line_point=2
        
    elif line=='원딜':
        line_point=3
        
    elif line=='서폿':
        line_point=4
        
    else:
        line_point=5
    return line_point
        
    
def ti(ti):
    
    if ti=='아1':
        point=1
    elif ti=='아2':
        point=1
    elif ti=='아3':
        point=1
    elif ti=='아4':
        point=1
    
    elif ti=='브1':
        point=3.5
    elif ti=='브2':
        point=3
    elif ti=='브3':
        point=2.5
    elif ti=='브4':
        point=2
    
    elif ti=='실1':
        point=5.5
    elif ti=='실2':
        point=5
    elif ti=='실3':
        point=4.5
    elif ti=='실4':
        point=4
    
    elif ti=='골1':
        point=7.5
    elif ti=='골2':
        point=7
    elif ti=='골3':
        point=6.5
    elif ti=='골4':
        point=6
    
    elif ti=='플1':
        point=12
    elif ti=='플2':
        point=11
    elif ti=='플3':
        point=9.5
    elif ti=='플4':
        point=8.5
    
    elif ti=='다1':
        point=19.5
    elif ti=='다2':
        point=17.5
    elif ti=='다3':
        point=15.5
    elif ti=='다4':
        point=14
    
    elif ti=='마스터':
        point=21.5
    elif ti=='그마':
        point=24.5
    elif ti=='챌':
        point=28.5
        
    else:
        point=0
    return point
```
